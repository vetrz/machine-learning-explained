```python
df = df.dropna()
df.isna().sum()
```

```python
df = df.drop_duplicates()
```

```python
for col in df.columns:
    print(f"{col}: {df[col].unique()}")
```

```python
df["income"] = df["income"].apply(lambda x: x.replace(".", ""))
df["income"].unique()
```

```python
df["occupation"].value_counts()
```

```python
df["native-country"].value_counts()
```

```python
df = df[(df["occupation"] != "?") & (df["native-country"] != "?")]
print(df.shape)
for col in df.columns:
    print(f"{col}: {df[col].unique()}")
```

```python
df = df.drop(columns=["education-num", "fnlwgt"])
df.columns
```

```python
df.to_csv("./data/processed/adult_processed.csv", index=False)
```
