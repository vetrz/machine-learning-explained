```python
import pandas as pd
```

```python
df_training = pd.read_csv("../data/raw/training_set.csv")
df_training.head()
```

```python
df_test = pd.read_csv("../data/base/test_set_features.csv")
df_test.head()
```

```python
df_training.info()
```

```python
print("Valores ausentes por coluna:")
print(df_training.isna().sum())
print(f"\nTotal de linhas com valores ausentes: {df_training.isna().any(axis=1).sum()}")
```

```python
df_training = df_training.dropna()
print(f"Shape após remoção de nulos: {df_training.shape}")
print("\nVerificação:")
print(df_training.isna().sum())
```

```python
num_duplicates = df_training.duplicated().sum()
print(f"Número de linhas duplicadas: {num_duplicates}")
```

```python
for col in df_training.columns:
    n_unique = df_training[col].nunique()
    print(f"{col} ({n_unique} valores únicos):")
    print(df_training[col].unique())
    print()
```

```python
df_training = df_training.drop(columns=["respondent_id"])
df_test = df_test.drop(columns=["respondent_id"])
```

```python
df_training.to_csv("../data/processed/training_set_cleaned.csv", index=False)
df_test.to_csv("../data/processed/test_set_cleaned.csv", index=False)
```
